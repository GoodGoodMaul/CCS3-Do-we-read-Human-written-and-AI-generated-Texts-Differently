```python
import pandas as pd
import statsmodels.formula.api as smf
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns


df_a = pd.read_excel('DatasetA.xlsx') 

df_b = pd.read_excel('DatasetB.xlsx') 

print("æ•°æ®åŠ è½½å®Œæˆã€‚")
```

## è·³è¯»ç‡

```python
print("========== RQ1: æ€»ä½“å·®å¼‚åˆ†æ (AI vs Human) ==========")

model_rq1_skip = smf.gee("Is_Skipped ~ Source + Number_of_units", 
                         data=df_a, 
                         groups=df_a["Participant"], 
                         family=sm.families.Binomial())
res_rq1_skip = model_rq1_skip.fit()
print("\n[RQ1 - è·³è¯»ç‡ç»“æœ]:")
print(res_rq1_skip.summary())
```

```python

model_rq1_fix = smf.mixedlm("Log_Avg_Fixation ~ Source + Number_of_units", 
                            data=df_b, 
                            groups=df_b["Participant"])
res_rq1_fix = model_rq1_fix.fit()
print("\n[RQ1 - æ³¨è§†æ—¶é•¿ç»“æœ]:")
print(res_rq1_fix.summary())
```

```python

model_rq1_reg = smf.mixedlm("Log_Reg_Path ~ Source + Number_of_units", 
                            data=df_b, 
                            groups=df_b["Participant"])
res_rq1_reg = model_rq1_reg.fit()
print("\n[RQ1 - å›è§†è·¯å¾„ç»“æœ]:")
print(res_rq1_reg.summary())
```

```python

model_rq1_pupil = smf.mixedlm("Pupil_Change ~ Source + Number_of_units", 
                              data=df_b, 
                              groups=df_b["Participant"])
res_rq1_pupil = model_rq1_pupil.fit()
print("\n[RQ1 - ç³å­”ç›´å¾„ç»“æœ]:")
print(res_rq1_pupil.summary())
```

```python
print("\n========== RQ2: äº¤äº’æ•ˆåº”åˆ†æ (Source x Word Category) ==========")

model_rq2_skip = smf.gee("Is_Skipped ~ Source * Word_category + Number_of_units", 
                         data=df_a, 
                         groups=df_a["Participant"], 
                         family=sm.families.Binomial())
res_rq2_skip = model_rq2_skip.fit()
print("\n[RQ2 - è·³è¯»ç‡äº¤äº’ç»“æœ]:")
print(res_rq2_skip.summary())
```

```python

model_rq2_fix = smf.mixedlm("Log_Avg_Fixation ~ Source * Word_category + Number_of_units", 
                            data=df_b, 
                            groups=df_b["Participant"])
res_rq2_fix = model_rq2_fix.fit()
print("\n[RQ2 - æ³¨è§†æ—¶é•¿äº¤äº’ç»“æœ]:")
print(res_rq2_fix.summary())
```

```python

model_rq2_reg = smf.mixedlm("Log_Reg_Path ~ Source * Word_category + Number_of_units", 
                            data=df_b, 
                            groups=df_b["Participant"])
res_rq2_reg = model_rq2_reg.fit()
print("\n[RQ2 - å›è§†è·¯å¾„äº¤äº’ç»“æœ]:")
print(res_rq2_reg.summary())
```

```python

model_rq2_pupil = smf.mixedlm("Pupil_Change ~ Source * Word_category + Number_of_units", 
                              data=df_b, 
                              groups=df_b["Participant"])
res_rq2_pupil = model_rq2_pupil.fit()
print("\n[RQ2 - ç³å­”äº¤äº’ç»“æœ]:")
print(res_rq2_pupil.summary())
```

```python
sns.set(style="whitegrid")
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

sns.pointplot(x="Word_category", y="Is_Skipped", hue="Source", 
              data=df_a, ax=axes[0,0], capsize=.1)
axes[0,0].set_title("Skipping Rate Interaction")

sns.pointplot(x="Word_category", y="Log_Avg_Fixation", hue="Source", 
              data=df_b, ax=axes[0,1], capsize=.1)
axes[0,1].set_title("Fixation Duration Interaction")

sns.pointplot(x="Word_category", y="Log_Reg_Path", hue="Source", 
              data=df_b, ax=axes[1,0], capsize=.1)
axes[1,0].set_title("Regression Path Interaction")

sns.pointplot(x="Word_category", y="Pupil_Change", hue="Source", 
              data=df_b, ax=axes[1,1], capsize=.1)
axes[1,1].set_title("Pupil Diameter Interaction")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import os


if not os.path.exists('figures'):
    os.makedirs('figures')


sns.set(style="whitegrid")

plt.figure(figsize=(5, 4)) 
sns.pointplot(x="Word_category", y="Is_Skipped", hue="Source", 
              data=df_a, 
              capsize=.1,           
              dodge=True,           
              markers=["o", "s"],   
              linestyles=["-", "--"], 
              palette="deep")       

plt.title("Interaction: Skipping Rate")
plt.ylabel("Skipping Probability")
plt.xlabel("Word Category")
plt.legend(title="Source", loc='best') 
plt.tight_layout()

plt.savefig('figures/interaction_skipping.pdf') 
plt.close() 
print("âœ… å·²ä¿å­˜: figures/interaction_skipping.pdf")



plt.figure(figsize=(5, 4))
sns.pointplot(x="Word_category", y="Log_Avg_Fixation", hue="Source", 
              data=df_b, 
              capsize=.1, dodge=True, 
              markers=["o", "s"], linestyles=["-", "--"],
              palette="deep")

plt.title("Interaction: Fixation Duration")
plt.ylabel("Log Fixation Duration")
plt.xlabel("Word Category")
plt.legend(title="Source", loc='best')
plt.tight_layout()

plt.savefig('figures/interaction_fixation.pdf')
plt.close()
print("âœ… å·²ä¿å­˜: figures/interaction_fixation.pdf")



plt.figure(figsize=(5, 4))
sns.pointplot(x="Word_category", y="Log_Reg_Path", hue="Source", 
              data=df_b, 
              capsize=.1, dodge=True, 
              markers=["o", "s"], linestyles=["-", "--"],
              palette="deep")

plt.title("Interaction: Regression Path")
plt.ylabel("Log Regression Path Duration")
plt.xlabel("Word Category")
plt.legend(title="Source", loc='best')
plt.tight_layout()

plt.savefig('figures/interaction_regression.pdf')
plt.close()
print("âœ… å·²ä¿å­˜: figures/interaction_regression.pdf")


plt.figure(figsize=(5, 4))
sns.pointplot(x="Word_category", y="Pupil_Change", hue="Source", 
              data=df_b, 
              capsize=.1, dodge=True, 
              markers=["o", "s"], linestyles=["-", "--"],
              palette="deep")

plt.title("Interaction: Pupil Diameter")
plt.ylabel("Pupil Diameter Change (mm)")
plt.xlabel("Word Category")
plt.legend(title="Source", loc='best')
plt.tight_layout()

plt.savefig('figures/interaction_pupil.pdf')
plt.close()
print("âœ… å·²ä¿å­˜: figures/interaction_pupil.pdf")
```
