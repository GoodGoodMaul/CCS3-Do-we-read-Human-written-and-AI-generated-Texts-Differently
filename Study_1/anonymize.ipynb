匿名化

```python
import pandas as pd

# ====== 参数 ======
input_path = "DatasetB.xlsx"      # 原始文件
output_path = "DatasetB_anon.xlsx"  # 匿名化后的文件
name_col = "Participant"       # 姓名列名
prefix = "P"                   # 匿名ID前缀，如 P001

# ====== 读取数据 ======
df = pd.read_excel(input_path)

# ====== 生成唯一姓名 -> 匿名ID 映射 ======
unique_names = df[name_col].unique()

id_map = {
    name: f"{prefix}{i+1:03d}"
    for i, name in enumerate(unique_names)
}

# ====== 应用映射 ======
df[name_col] = df[name_col].map(id_map)

# ====== 保存结果 ======
df.to_excel(output_path, index=False)

print("匿名化完成！")
print("映射示例（前5个）：")
for k, v in list(id_map.items())[:5]:
    print(f"{k} -> {v}")
```
